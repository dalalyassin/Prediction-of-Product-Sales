```python
#nominal, ordinal, numerical
num_cols= df.select_dtypes(include=['number']).columns
num_cols
```

```python
ord_cols = ['Item_Fat_Content', 'Outlet_Identifier', 'Outlet_Size']
#ethod does not support multiple arguments. Instead, it expects a single argument (the label(s) to drop).|
to_drop = ord_cols + ['Item_Identifier']
cat_cols = df.select_dtypes(include=['object']).columns.drop(to_drop)

#Item_Identifier, Item_Type . Outlet_Location_Type,Outlet_Type, nominal
#Item_Fat_Content, Outlet_Identifier,Outlet_Size  ordinal
```

```python
#target
y = df['Item_Outlet_Sales']
# The features we will use to make the prediction
X = df.drop(columns=['Item_Identifier'])
```

```python
from sklearn.model_selection import train_test_split

# Train test split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
X_train[num_cols].isna().sum()
```

```python
#start with num cols
# instantiate preprocessors
from sklearn.pipeline import make_pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler


impute_median = SimpleImputer(strategy='median')
scaler = StandardScaler()

num_pipe = make_pipeline(impute_median, scaler)
num_pipe
```

```python
# Fit the pipeline on the training data
num_pipe.fit(X_train[num_cols])
```

```python
X_train_num_trans = num_pipe.transform(X_train[num_cols])
X_train_num_trans = num_pipe.transform(X_test[num_cols])
```

```python
num_tuple = ('numeric', num_pipe, num_cols)
```

```python
#categorical
from sklearn.preprocessing import OneHotEncoder
impute_na = SimpleImputer(strategy='constant', fill_value = "MISSING")
#encode the features
cat_encoder = OneHotEncoder(sparse_output=False, handle_unknown='ignore')
```

```python
cat_pipe = make_pipeline(impute_na, cat_encoder)
cat_pipe
```

```python
cat_pipe.fit(X_train[cat_cols])
```

```python
X_train_cat_trans = cat_pipe.transform(X_train[cat_cols])
X_train_cat_trans = cat_pipe.transform(X_test[cat_cols])
```

```python
cat_tuple = ('object', cat_pipe, cat_cols)
```

```python

#for nominal feautes
df[ord_cols].head()
```

```python
qual_cond_order1 = ['Regular', 'Low Fat']
qual_cond_order2 = ['OUT049', 'OUT018', 'OUT010', 'OUT013', 'OUT027',
                   'OUT045', 'OUT017', 'OUT046', 'OUT035', 'OUT019']
qual_cond_order3 = ['Medium', 'High', 'Small']
```

```python
from sklearn.preprocessing import OrdinalEncoder
impute_na_ord = SimpleImputer(strategy='constant', fill_value='NA')
ordinal_category_orders = [qual_cond_order1, qual_cond_order2,qual_cond_order3]
ord_encoder = OrdinalEncoder(categories=ordinal_category_orders, handle_unknown='error')
scaler_ord = StandardScaler()
# Making an ord_pipe
ord_pipe = make_pipeline(impute_na_ord, ord_encoder, scaler_ord)
ord_pipe
```

```python
# Fit the pipeline on the training data
ord_pipe.fit(X_train[ord_cols])
```

```python
from sklearn import set_config
set_config(transform_output='pandas')
# Transform the training and test data using the pipeline
X_train_ord_tf = ord_pipe.transform(X_train[ord_cols])  # Transform the training data
X_train_ord_tf = ord_pipe.transform(X_test[ord_cols])
# Display the first few rows of the transformed training data
```

```python
ord_tuple = ('ordinal', ord_pipe, ord_cols)
```

```python
from sklearn.compose import ColumnTransformer

#- Create a column transformer object that encompasses the 3 preprocessing pipelines from the previous assignment.
col_transformer = ColumnTransformer([num_tuple, cat_tuple , ord_tuple],
                                    verbose_feature_names_out=False)
col_transformer.fit(X_train)
```

```python
# Transform the training data
X_train_processed = col_transformer.transform(X_train)
# Transform the testing data
X_test_processed = col_transformer.transform(X_test)
X_train_processed.head()
```
