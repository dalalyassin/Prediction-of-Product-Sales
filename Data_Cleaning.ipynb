```python
null_sums = df.isna().sum()
null_sums
```

```python
item=df.groupby('Item_Type')['Item_Weight'].mean()
item
```

```python
for item_type in df['Item_Type'].unique():
    missing_count = df[df['Item_Type'] == item_type]['Item_Weight'].isna().sum()

    print(f"Missing values for '{item_type}' is: {missing_count}")
```

```python
df['Item_Weight'] = df.apply(
    lambda row: item[row['Item_Type']] if pd.isna(row['Item_Weight']) else row['Item_Weight'],
    axis=1)
```

```python
df['Item_Weight'].isna().sum()
```

```python
size=df['Outlet_Size'].mode().iloc[0]
#iloc[0] if we have more than one mode and need to choose one
size
```

```python
df['Outlet_Size'] = df['Outlet_Size'].fillna(size)
```

```python
df['Outlet_Size'].isna().sum()
```

Histograms to view the distributions of numerical features in my dataset.


```python
numerical_features = df.select_dtypes(include=['number']).columns
for i in numerical_features:
     plt.figure(figsize=(8, 6)) #.figure and not .show --> .figure object in matplotlib to draw x axis and y axis without specifications.>> bar or box>>
     sns.histplot(data=df, x=i)
     plt.title(f'Distribution of {i}')
     plt.xlabel(i)
     plt.show()
```

Boxplots to view statistical summaries of numerical features

```python
numerical_features = df.select_dtypes(include=['number']).columns
for i in numerical_features:
     plt.figure(figsize=(10, 8))
     ax = sns.boxplot(data=df, y=i)
     plt.title(f'Distribution of {i}')
     plt.xlabel(i)
     plt.show()
     ax
```

Countplots to view the frequency of each class of categorial feature

```python
categorical_features = df.select_dtypes(include=['object']).columns
for i in categorical_features:
     plt.figure(figsize=(8, 6))
     ax = sns.boxplot(data=df, y=i)
     plt.title(f'Distribution of {i}')
     plt.xlabel(i)
     plt.show()
     ax
```
